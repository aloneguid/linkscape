```python
from linkscape.network import Network
```

```python
# read vehicles.csv
import pandas as pd
vehicles = pd.read_csv("vehicles.csv")
vehicles
```

```python
g = Network(800, 600, collision_radius=25)

def make_nodes(col_name: str, colour: str):
    values = vehicles[col_name].unique().tolist()
    for value in values:
        n = g.node(value, value)
        n.color = colour
        n.radius = 5
        n.stroke_width = 0

make_nodes("make", None)
make_nodes("transmission", "red")
make_nodes("fuel", "blue")

# group by make and transmission
grouped = vehicles.groupby(["make", "transmission"]).size().reset_index(name="counts")

for i, row in grouped.iterrows():
    g.edge(row["make"], row["transmission"])
    
# group by transmission and fuel
grouped = vehicles.groupby(["transmission", "fuel"]).size().reset_index(name="counts")

for i, row in grouped.iterrows():
    g.edge(row["transmission"], row["fuel"])

g
```
