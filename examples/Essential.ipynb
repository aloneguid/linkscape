# Essential examples

```python
from linkscape.network import Network
```

## Nodes

To start with, let's create a basic network with 3 nodes and 2 edges. First, create an instance of the Network class, specifying canvas size. Then, add nodes to the network.

```python
g = Network(800, 200)

g.node("1")
g.node("2")
g.node("3")

g
```

`.node()` method returns `Node` object, which can be used to customize node properties.

```python
g = Network(800, 100)
node = g.node("4")
node.label = "Node 4"
node.radius = 5
node.tooltip = "This is node 4"
node.color = "red"
node.stroke = "black"
node.stroke_width = 2
g
```

## Edges

To add edges to the network, use the `edge()` method. It takes two node ids as arguments, source node id and target node id.

```python
g = Network(800, 200)
g.node("1")
g.node("2")
g.edge("1", "2")
g
```

`.edge()` method returns `Edge` object, which can be used to customize edge properties.

```python
g.node("3")
edge = g.edge("2", "3")
edge.color = "black"
edge.opacity = 0.8
edge.width = 2
edge.label = "Edge 2-3"
edge.force = 0.01
g
```
